## Isolation Forests on Subsetted Data
Isolation Forests do not suffer from the curse of dimensionality so we used it to detect anomalies on subsetted data, which still has a large number of features. First we identified a set of outliers based every single feature in the data. The plots illustrate the positions of the outliers in 2-D and 3-D PCA space.

```python
# note for when we combine this:
# everything above this is directly copied from ryan-unsupervised

# isolation forest documentation: https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.IsolationForest.html
## max_features: int or float, optional (default=1.0)
## bootstrap: bool, optional (default=False)

from sklearn.ensemble import IsolationForest as lf
lf = lf(max_features = len(data.columns))
pred=lf.fit_predict(data)

plot_2d(dim_reduct_for_plot(data),pred).show()
plot_3d(dim_reduct_for_plot(data, dim = 3),pred).show()
```

We also ran an Isolation Forest on each feature individually. The first 2-D and 3-D plots show the number of anomalous features in the PCA space. The histogram and line plot illustrate how many outlier features the samples have. This is useful because it shows to what degree the samples are an outlier. However, a sample with many outliers may be an extreme, not an outlier. We arbitrarily chose an outlier cutoff of 100 or more anomalous features (about a quarter of the features). The second set of 2-D and 3-D plots show the points that are above this cutoff, and therefore considered anomalies.

```python
from sklearn.ensemble import IsolationForest
data2 = data.copy()
data2["anomalies"] = 0
for i in data2.columns[:-1]:
    X = data2[i] #index column
    temp = (X.to_numpy()).reshape(-1, 1)
    lf = IsolationForest(bootstrap = False) #make lf object
    pred2 = lf.fit_predict(temp) #run lf on column
    pred2[pred2 == 1]=0
    pred2[pred2 == -1]=1
    data2["anomalies"] = data2["anomalies"] + pred2
plot_2d(dim_reduct_for_plot(data),data2["anomalies"]).show()
plot_3d(dim_reduct_for_plot(data, dim = 3),data2["anomalies"]).show()
#data2["anomalies"]
```

## Local Outlier Factor
LOcal Outlier Factor (LOF) anomaly detection uses density to find outliers. Not knowing which combination of LOF metric, K nearest neighbor, and algrithen would work best, we used every combination of the sklearn options, as well as a range of KNN values. Some combinations were incompatible and therefore could not be used. We used the results of each to vote on which samples would be considered a LOF outlier.

```python
# LOF documentation: https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.LocalOutlierFactor.html
# Voting Reference: https://stackoverflow.com/questions/55749114/how-to-do-a-majority-voting-on-columns-in-pandas
from sklearn.neighbors import LocalOutlierFactor
temp = [0]*len(data)
LOFvoting = pd.DataFrame({"temp": temp})
for h in list(range(5,30,5)):
    for i in ["cityblock", "cosine", "euclidean", "l1", "l2", "manhattan", "minkowski"]:
        for j in ["auto", "ball_tree", "kd_tree", "brute"]:
            try:
                LOF = LocalOutlierFactor(algorithm = j, metric = i, n_neighbors = h)
                key = str(h) + i + j
                LOFvoting[key] = LOF.fit_predict(data)
            except:
                print("Error:", str(h), i, j)
#df['majority'] = df.mode(axis=1)[0] #voting

anomaly_voting["LOF"] = (LOFvoting.mode(axis=1)[0]).to_numpy()
plot_2d(dim_reduct_for_plot(data),anomaly_voting["LOF"]).show()
plot_3d(dim_reduct_for_plot(data, dim = 3),anomaly_voting["LOF"]).show()
```
